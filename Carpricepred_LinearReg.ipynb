```python
raw_data = pd.read_csv(r'C:\Users\ISHAN\Downloads\1.04.+Real-life+example.csv')
raw_data
```

```python
#we want to predict the price of a used car depending on its specifications.
#we understand that the brand,mileage,engineV and year are the most important regressors.
```

```python
#we now drop the unwanted columns from the csv file
#this is also known as data cleaning
data = raw_data.drop(['Model'],axis = 1)
#here axis = 1 means column and axis = 0 means a row
```

```python
data_no_mv.describe(include='all')
#here we observe that the totl count has reduced to 4025 from 4345 after deleting the rows with missing values.
```

```python
#we use the quantile function to deal with the outliers.
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include = 'all')
```

```python
#now we note that the max price has dropped from 300000 to 129222 which is much closer and acceptable than the previous value.
```

```python
sns.displot(data_1['Price'])
#we can see below that the graph is different from the one above.
```

```python
#we notice that there is a similiar observation of the pdf for mileage so we treat it in a similiar way as price.
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]
sns.displot(data_2['Mileage'])
```

```python
#after observing some unusual disperencies in the EngineV column we decide to only take values up till 6.5 for our predictions.
data_3 = data_2[data_2['EngineV']<6.5]
data_3.describe(include = 'all')
```

```python
sns.displot(data_3['EngineV'])
```

```python
#calculation for year to filter out vintage cars
sns.displot(data_3['Year'])
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
sns.displot(data_4['Year'])
```

```python
#now we have sucessfully cleaned the data and now we want to store it while the indexes have been reset
#we reset the indexes because they are still of the same value that they were before the cleaning was done
#we do that using
cleaned_data = data_4.reset_index(drop=True)
cleaned_data
#we used drop to completely forget about the old indexes
```

```python
cleaned_data.describe(include = 'all')
```

```python
#Checking the OLS Assumptions
f, (ax1,ax2,ax3) = plt.subplots(1,3,sharey = True,figsize = (15,3))
ax1.scatter(cleaned_data['Year'],cleaned_data['Price'])
ax1.set_title('Price and Year')
ax2.scatter(cleaned_data['EngineV'],cleaned_data['Price'])
ax2.set_title('Price and Engine Volume')
ax3.scatter(cleaned_data['Mileage'],cleaned_data['Price'])
ax1.set_title('Price and Mileage')
```

```python
#Since we can conclude from the above graphs that these distributions are not linear we cannot use the linear transformation 
#for them. we use the log transformation when dealing with exponential distributions
log_price = np.log(cleaned_data['Price'])
log_price
#we use np.log() to get the log values of a number or an array of numbers.It is an operator that is present in the pandas lib.
```

```python
sns.displot(cleaned_data['Price'])
```

```python
#now we add the log values of price into our datasheet as a new entry column
cleaned_data['Log_Price'] = log_price
cleaned_data
```

```python
#now we plot the graph again using log_prices
f, (ax1,ax2,ax3) = plt.subplots(1,3,sharey = True,figsize = (15,3))
ax1.scatter(cleaned_data['Year'],cleaned_data['Log_Price'])
ax1.set_title('Log_Price and Year')
ax2.scatter(cleaned_data['EngineV'],cleaned_data['Log_Price'])
ax2.set_title('Log_Price and Engine Volume')
ax3.scatter(cleaned_data['Mileage'],cleaned_data['Log_Price'])
ax3.set_title('Log_Price and Mileage')
```

```python
#now we can se that there is a pattern in all the graphs and that they fit the data very well.
#we can now drop the original price column as it is not gonna be used
cleaned_data = cleaned_data.drop(['Price'],axis = 1)
cleaned_data
```

```python
#now we have to perform the multicollinearity test
#multicol. is a regression assumption in which 2 or more variables have a strong correlation amongst themselves.
#sklearn does not have a built in function for testing this so we use statsmodel for this test.
#one of the best ways to test for multicol. is VIF(variable inflation factor)
#VIF produces a measure that estimates how much larger the square root of standard error of an estimate is compared to a 
#situation where the variable was completely un correlated to another predictor 
from statsmodels.stats.outliers_influence import variance_inflation_factor
#we have 3 continous variables year,enginev and mileage
variables = cleaned_data[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
```

```python
vif
```

```python
#when VIF = 1 it means there is no multicollinearity
#when 1<VIF<5 it is an acceptable value
#anything above 5 has strong multicollinearity 
#we rarely find a data so perfect that there is not a single variable which is not multi collinear
#here "Year" has the strongest multi collinearity so we drop it from our dataset.
#cleaned_data_noMC = cleaned_data.drop(['Year'],axis = 1)
cleaned_data_noMC=cleaned_data
cleaned_data_noMC
```

```python
##Now we create dummy variables
#a dummy variable is a variabe in regression which is used to include categorical data.
#pandas has a useful function called "pd.get_dummies". It automatically spots all categorical variables and creates dummies.
#if we have "n" categories for a feature we have to create "n-1" dummies.
data_dum = pd.get_dummies(cleaned_data_noMC, drop_first = True)
data_dum.head()
```

```python
#now we rearrange the columns a bit
data_dum.columns.values
```

```python
cols = [ 'Log_Price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
preprocessed_data = data_dum[cols]
preprocessed_data
```

```python
#Linear regression model
#declaring the inputs and the targets
target = preprocessed_data['Log_Price']
inputs = preprocessed_data.drop(['Log_Price'],axis = 1)
```

```python
#scaling of data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
#it is not usually reccomended to standardise dummy variables
#scaling has no effect on the predictive power of dummies however once they are scaled they are no longer dummies
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs_scaled,target,test_size=0.2,random_state=365)
```

```python
#creating the regression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#A simple way to check the final result is to plot the predicted values against the observed values
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets(y_train)',size = 18)
plt.ylabel('Predictions(y_hat)',size = 18)
plt.show()
```

```python
#Residual = Difference between targets and the predictions
sns.displot(y_train - y_hat)
plt.title("Residual PDF",size = 15)
```

```python
reg.score(x_train,y_train)
```

```python
#finding weights and bias
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
#weights interpretation
#a positive weight shows that as the feature increases in value the log_price and price also increase
#a negative weight shows that as the feature increases in value the log_price and price decrease.
#we need to check for the benchmark feature in brand
cleaned_data['Brand'].unique()
```

```python
#from the above we understand that audi is the benchmark value brand
#weight interpretation for dummy variables
#a positive weight tells us that the respective brand which is being compared to audi is more expensive than audi
#a negative weight tells us that the respective brand which is being compared to audi is cheaper as compared to audi.
```

```python
#bigger the weight higher the impact.
#lower the weight lower the impact.
#dummies are only compared to the benchmark and not compared to continous variables.
```

```python
#Testing the model
y_hat_test = reg.predict(x_test)
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('Targets(y_test)',size = 18)
plt.ylabel('Predictions(y_hat_test)',size = 18)
plt.show()
```

```python
#by observing the model we understand that there is a larger concentration of points close to 45deg line at the upper right end
#of the scatter plot.
#this shows that our model is very good at predicting expensive cars but on the opposite end there is a larger spread of points
#which shows that it is not comparatively good at predicting cheaper cars.
#we use alpha to make the points opaque so that the data is much more visible to users and insights gathering.
#it takes values from 0 to 1 with 1 being the default value.
#the above graph represents a heat map where darker the color is the higher the concentration of points is.
```

```python
#to get a perfomance prediction we use the following
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf
```

```python
#now these prices are in their log form and since we can take exponentials of the log form we can get the original prices.
#for this we use np.exp(x) and it returns the exponential of x(e to the power of x)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test
```

```python
#y test with original indices dropped
y_test = y_test.reset_index(drop = True)
y_test
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
#to show the difference between predictions and target we create another column called residual
df_pf['Residual'] = df_pf['Predictions'] - df_pf['Target']
df_pf
```

```python
#the ols framework is based on minimising SSE(sum of squared errors) and that is obtained by minimising sm of sqaured residuals.
df_pf['% diff'] = np.absolute(df_pf['Residual']/df_pf['Target'] * 100)
df_pf
```
